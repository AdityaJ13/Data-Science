```python
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
#inaugural.raw('1941-Roosevelt.txt')
#inaugural.raw('1961-Kennedy.txt')
#inaugural.raw('1973-Nixon.txt')
import string
import pandas as pd
```

```python
rsvlt = inaugural.raw('1941-Roosevelt.txt')
kndy = inaugural.raw('1961-Kennedy.txt')
nixon = inaugural.raw('1973-Nixon.txt')
```

# 2.1) Find the number of characters, words and sentences for the mentioned documents. (Hint: use .words(), .raw(), .sent() for extracting counts)

### Counting no of characters

```python
print('No of characters in Roosevelts speech:', len(rsvlt))
print('No of characters is Kennedys speech:', len(kndy))
print('No of characters in Nixons speech:', len(nixon))
```

### Couting no of words : using words()

```python
print('No of words in Roosevelts speech is:', 
      len(inaugural.words(fileids='1941-Roosevelt.txt')))
print('No of words in Kenedys speech is:', 
      len(inaugural.words(fileids='1961-Kennedy.txt')))
print('No of words in Nixon speech is:', 
      len(inaugural.words(fileids='1973-Nixon.txt')))
```

### Counting no of words: using split()

```python
print('No of words in Roosevelt speech:', 
      len(inaugural.raw(fileids = '1941-Roosevelt.txt').split()))
print('No of words in Kenedys speech:', 
      len(inaugural.raw(fileids = '1961-Kennedy.txt').split()))
print('No of words in Nixons speech:', 
      len(inaugural.raw(fileids = '1973-Nixon.txt').split()))
```

### Counting no of sentences

```python
print('No of sentences in Roosevelt speech:',
      len(inaugural.sents(fileids='1941-Roosevelt.txt')))
print('No of sentences in Kennedys speech:',
      len(inaugural.sents(fileids = '1961-Kennedy.txt')))
print('No of sentences in Nixons speech:',
      len(inaugural.sents(fileids = '1973-Nixon.txt')))
```

# 2.2) Remove all the stopwords from the three speeches.

### Removing Stop Words

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation) + ['--']
all_words_rsvlt = (w.lower() for w in inaugural.words(fileids='1941-Roosevelt.txt'))
all_words_clean_rsvlt = [word for word in all_words_rsvlt if word not in stopwords]
```

#### Before cleaning

```python
inaugural.raw('1941-Roosevelt.txt')
```

#### After cleaning

```python
' '.join(all_words_clean_rsvlt)
```

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation) + ['--']
all_words_kndy = (w.lower() for w in inaugural.words(fileids='1961-Kennedy.txt'))
all_words_clean_kndy = [word for word in all_words_kndy if word not in stopwords]
```

#### Before cleaning

```python
inaugural.raw('1961-Kennedy.txt')
```

#### After cleaning

```python
' '.join(all_words_clean_kndy)
```

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation) + ['--']
all_words_nixon = (w.lower() for w in inaugural.words(fileids='1973-Nixon.txt'))
all_words_clean_nixon = [word for word in all_words_nixon if word not in stopwords]
```

#### Before removing stop words

```python
inaugural.raw('1973-Nixon.txt')
```

#### After cleaning

```python
' '.join(all_words_clean_nixon)
```

# 2.3) Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords)

### Most commonly occuring words in each of the speeches

```python
rsvlt_freq = nltk.FreqDist(all_words_clean_rsvlt)
pd.DataFrame(rsvlt_freq.most_common(3), columns = ['Word','Frequency'])
```

```python
kndy_freq = nltk.FreqDist(all_words_clean_kndy)
pd.DataFrame(kndy_freq.most_common(3), columns = ['Word','Frequency'])
```

```python
nixon_freq = nltk.FreqDist(all_words_clean_nixon)
pd.DataFrame(nixon_freq.most_common(3), columns = ['Word','Frequency'])
```

# 2.4) Plot the word cloud of each of the three speeches. (after removing the stopwords)

### Word cloud

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc_rsvlt = ' '.join(all_words_clean_rsvlt)
wc_kndy = ' '.join(all_words_clean_kndy)
wc_nixon = ' '.join(all_words_clean_nixon)
```

```python
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='white', 
                min_font_size = 10, random_state=100).generate(wc_rsvlt) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for ROOSEVELT'S SPEECH (after cleaning)!!")
plt.show()
```

```python
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='white', 
                min_font_size = 10, random_state=100).generate(wc_kndy) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for KENNEDY'S SPEECH (after cleaning)!!")
plt.show()
```

```python
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='white', 
                min_font_size = 10, random_state=100).generate(wc_nixon) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for NIXON'S SPEECH (after cleaning)!!")
plt.show()
```
